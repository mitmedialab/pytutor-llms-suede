# JSON Schema Structured Output Examples
One cell per provider/model setup from benchmark cases.

```python
import json
import patch
from release import Provider
from release.providers.base import ModelMetadata
from release.providers.openai import OpenAIProvider
from release.providers.google import GoogleProvider
from release.providers.anthropic import AnthropicProvider
from release.providers.openrouter import OpenRouterProvider
```

```python
SCHEMA = {
    "type": "object",
    "properties": {
        "question": {
            "type": "string",
            "description": "The question component of a multiple choice question.",
        },
        "options": {
            "type": "array",
            "description": "The options in a multiple choice question (typically a total of 4).",
            "items": {
                "type": "object",
                "properties": {
                    "content": {
                        "type": "string",
                        "description": "The content of the option in a multiple choice question.",
                    },
                    "correct": {
                        "type": "boolean",
                        "description": "Whether the option is correct or not. There should be only one correct option.",
                    },
                    "explanation": {
                        "type": "string",
                        "description": "An explanation of why the option is correct or incorrect.",
                    },
                },
                "required": ["content", "correct", "explanation"],
                "additionalProperties": False,
            },
        },
    },
    "required": ["question", "options"],
    "additionalProperties": False,
    "title": "GeneratedMultipleChoiceQuestion",
    "$schema": "http://json-schema.org/draft-07/schema#",
}
```

```python
async def run_schema_case(name: str, provider: Provider, model: str, model_metadata: list[ModelMetadata]) -> None:
    stream = await Provider.SchemaStream.Select(
        Provider.SchemaStream.Request(
            model=model,
            model_metadata=model_metadata,
            schema=json.dumps(SCHEMA),
            messages=[
                {'role': 'system', 'content': 'Return valid structured data.'},
                {'role': 'user', 'content': 'Quiz me on python loops.'},
            ],
        ),
        provider,
    )
    final_model = None
    print(f'[{name}] model={model}')
    async for event in stream:
        if event.type == 'error':
            print(f'[Error: {event.payload}]')
            return
        final_model = event.payload.current
        print(f'partial[{event.payload.i}]: {final_model.model_dump()}')
    if final_model is not None:
        print('final:')
        print(final_model.model_dump_json(indent=2))
```

```python
await run_schema_case('openai 4o', OpenAIProvider(), 'gpt-4o', [OpenAIProvider.ModelMetadata()])
```

```python
await run_schema_case('openai 5.2', OpenAIProvider(), 'gpt-5.2-2025-12-11', [OpenAIProvider.ModelMetadata()])
```

```python
await run_schema_case('google', GoogleProvider(), 'gemini-3-flash-preview', [GoogleProvider.ModelMetadata(thinking_level='minimal')])
```

```python
await run_schema_case('anthropic', AnthropicProvider(), 'claude-sonnet-4-5-20250929', [AnthropicProvider.ModelMetadata()])
```
