# Text Streaming Examples
One cell per provider/model setup from benchmark cases.

```python
import patch
from release import Provider
from release.providers.base import ModelMetadata
from release.providers.openai import OpenAIProvider
from release.providers.google import GoogleProvider
from release.providers.anthropic import AnthropicProvider
from release.providers.openrouter import OpenRouterProvider
```

```python
async def run_text_case(name: str, provider: Provider, model: str, model_metadata: list[ModelMetadata]) -> None:
    stream = await Provider.TextStream.Select(
        Provider.TextStream.Request(
            model=model,
            model_metadata=model_metadata,
            messages=[
                {'role': 'system', 'content': 'You are concise.'},
                {'role': 'user', 'content': 'In one sentence, explain what streaming responses are.'},
            ],
        ),
        provider,
    )
    print(f'[{name}] model={model}')
    async for event in stream:
        if event.type == 'error':
            print(f'[Error: {event.payload}]')
            return
        print(event.payload.delta, end='', flush=True)
    print()
```

```python
await run_text_case('openai 4o', OpenAIProvider(), 'gpt-4o', [OpenAIProvider.ModelMetadata()])
```

```python
await run_text_case('openai 5.2', OpenAIProvider(), 'gpt-5.2-2025-12-11', [OpenAIProvider.ModelMetadata()])
```

```python
await run_text_case('google', GoogleProvider(), 'gemini-3-flash-preview', [GoogleProvider.ModelMetadata(thinking_level='minimal')])
```

```python
await run_text_case('anthropic', AnthropicProvider(), 'claude-sonnet-4-5-20250929', [AnthropicProvider.ModelMetadata()])
```

```python
await run_text_case('openrouter', OpenRouterProvider(), 'moonshotai/kimi-k2.5', [OpenRouterProvider.ModelMetadata(reasoning_enabled=False)])
```
